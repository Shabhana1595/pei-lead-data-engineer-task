```python
%sql
-- ============================================================
-- 1. Top 10 Customers by Sales
-- ============================================================

SELECT *
FROM workspace.pei_assignment.gold_customer_summary
ORDER BY total_sales DESC
LIMIT 10;
```

```python
%sql
-- ============================================================
-- 2. Top 5 Product Categories by Profit
-- ============================================================

SELECT *
FROM workspace.pei_assignment.gold_category_summary
ORDER BY total_profit DESC
LIMIT 5;
```

```python
%sql
-- ============================================================
-- 3. State-wise Sales
-- ============================================================

SELECT *
FROM workspace.pei_assignment.gold_state_summary
ORDER BY total_sales DESC;
```

```python
%sql
-- ============================================================
-- 4. Yearly Sales Trend
-- ============================================================

SELECT *
FROM workspace.pei_assignment.gold_yearly_summary
ORDER BY order_year;
```

```python
%sql
-- ============================================================
-- 5. Top 10 Most Profitable Products
-- ============================================================

SELECT
    product_name,
    category,
    ROUND(SUM(profit),2) AS total_profit
FROM workspace.pei_assignment.order_enriched
GROUP BY product_name, category
ORDER BY total_profit DESC
LIMIT 10;
```

```python
%sql
-- ============================================================
-- 6. Average Discount by Category
-- ============================================================

SELECT
    category,
    ROUND(AVG(discount),2) AS avg_discount
FROM workspace.pei_assignment.order_enriched
GROUP BY category
ORDER BY avg_discount DESC;
```
